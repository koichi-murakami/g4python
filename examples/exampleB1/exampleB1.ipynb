```python
from geant4 import *
from geant4.utils import ParticleGun
import geomB1
```

```python
# User Action Initialization
class AppBuilder(G4VUserActionInitialization):
    def Build(self):
        # setup particle gun
        global particle_gun
        particle_gun = ParticleGun()
        self.SetUserAction(particle_gun)

        # setup particle gun
        pg = particle_gun.GetGun()
        pg.SetParticleByName("e-")
        pg.SetParticleEnergy(100.*MeV)
        pg.SetParticleMomentumDirection(G4ThreeVector(0.,0.,1.))
        pg.SetParticlePosition(G4ThreeVector(0.,0.,-16.)*cm)
```

```python
# set detectot construction
global b1geom
b1geom = geomB1.geomB1()
gRunManager.SetUserInitialization(b1geom)


# set physics list
global physics_list
physics_list = FTFP_BERT()
gRunManager.SetUserInitialization(physics_list)

# user action initialization
global app_builder
app_builder = AppBuilder()
gRunManager.SetUserInitialization(app_builder)
```

```python
gRunManager.Initialize()
gRunManager.BeamOn(0)
```
